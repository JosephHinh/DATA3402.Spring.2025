```python
# Create "Counter" class
class Counter:
    def __init__(self, max_value):
        self.max_value = max_value
        self.value = 0

    # Increases the value by 1 unless the counter has reached the max.
    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        else:
            print("Error: Counter has reached its maximum value.")
    
    # Resets the counter to 0.
    def reset(self):
        self.value = 0
    
    def get_value(self):
        return self.value

# Test the code out:
counter = Counter(5)
counter.increment()
counter.increment()
print(counter.get_value())  # Expected output: "2"
counter.increment()
counter.increment()
counter.increment()
counter.increment()  # Expected Output: "Error: Counter has reached the maximum value."
counter.reset()
print(counter.get_value())  # Expected output: "0"
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
# Add the new "overlaps" function to every shape class.
import math
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    
    @abstractmethod
    def perimeter(self):
        pass
    
    @abstractmethod
    def get_boundary_points(self):
        pass
    
    @abstractmethod
    def is_inside(self, x, y):
        pass
    
    @abstractmethod
    def overlaps(self, other):
        pass

class Circle(Shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y
    
    def get_radius(self):
        return self.__radius
    
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def area(self):
        return math.pi * self.__radius ** 2
    
    def perimeter(self):
        return 2 * math.pi * self.__radius
    
    def get_boundary_points(self):
        points = []
        for i in range(16):
            angle = (2 * math.pi / 16) * i
            x = round(self.__x + self.__radius * math.cos(angle), 2)
            y = round(self.__y + self.__radius * math.sin(angle), 2)
            points.append((x, y))
        return points
    
    def is_inside(self, x, y):
        return (x - self.__x) ** 2 + (y - self.__y) ** 2 <= self.__radius ** 2
    
    def overlaps(self, other):
        if isinstance(other, Circle):
            distance = math.sqrt((self.__x - other.__x) ** 2 + (self.__y - other.__y) ** 2)
            return distance < (self.__radius + other.__radius)
        return False

class Rectangle(Shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
    
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * (self.__length + self.__width)
    
    def get_boundary_points(self):
        points = [
            (round(self.__x, 2), round(self.__y, 2)),
            (round(self.__x + self.__length, 2), round(self.__y, 2)),
            (round(self.__x + self.__length, 2), round(self.__y + self.__width, 2)),
            (round(self.__x, 2), round(self.__y + self.__width, 2))
        ]
        return points
    
    def is_inside(self, x, y):
        return self.__x <= x <= self.__x + self.__length and self.__y <= y <= self.__y + self.__width
    
    def overlaps(self, other):
        if isinstance(other, Rectangle):
            return not (self.__x + self.__length < other.__x or other.__x + other.__length < self.__x or
                        self.__y + self.__width < other.__y or other.__y + other.__width < self.__y)
        return False

class Triangle(Shape):
    def __init__(self, side1, side2, side3, x, y):
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3
        self.__x = x
        self.__y = y
    
    def get_side1(self):
        return self.__side1
    
    def get_side2(self):
        return self.__side2
    
    def get_side3(self):
        return self.__side3
    
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def area(self):
        s = (self.__side1 + self.__side2 + self.__side3) / 2
        return math.sqrt(s * (s - self.__side1) * (s - self.__side2) * (s - self.__side3))
    
    def perimeter(self):
        return self.__side1 + self.__side2 + self.__side3
    
    def get_boundary_points(self):
        return [
            (round(self.__x, 2), round(self.__y, 2)),
            (round(self.__x + self.__side1, 2), round(self.__y, 2)),
            (round(self.__x + self.__side1 / 2, 2), round(self.__y + self.__side2, 2))
        ]
    
    def is_inside(self, x, y):
        x1, y1 = self.__x, self.__y
        x2, y2 = self.__x + self.__side1, self.__y
        x3, y3 = self.__x + self.__side1 / 2, self.__y + self.__side2
        
        def area(x1, y1, x2, y2, x3, y3):
            return abs((x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2)) / 2.0)
        
        A = area(x1, y1, x2, y2, x3, y3)
        A1 = area(x, y, x2, y2, x3, y3)
        A2 = area(x1, y1, x, y, x3, y3)
        A3 = area(x1, y1, x2, y2, x, y)
        
        return abs(A - (A1 + A2 + A3)) < 1e-9
    
    def overlaps(self, other):
        for point in other.get_boundary_points():
            if self.is_inside(*point):
                return True
        return False

# Test the new overlaps function for each class.
triangle = Triangle(6, 8, 10, 3, 6)
triangle2 = Triangle(3, 4, 5, 3, 6)
print(f"Do these two triangles overlap?: {triangle.overlaps(triangle2)}")

circle = Circle(7, 1, 0)
circle2 = Circle(7, 2, 0)
print(f"Do these two circles overlap?: {circle.overlaps(circle2)}")

rectangle = Rectangle(12, 5, 1, 1)
rectangle2 = Rectangle(12, 5, 30, 20)
print(f"Do these two rectangles overlap?: {rectangle.overlaps(rectangle2)}")
```
